```python
import pandas as pd

data = {
    "Name" : ["Sam", "Jhon", "Bob"],
    "Age" : [20, 65, 37]
}

df = pd.DataFrame(data, index=["Person 1", "Person 2", "Person 3"])

# Add new column
df["Job"] = ["SE", "QA", "N/A"]

# Add new row
new_row = pd.DataFrame([{"Name" : "Kamal", "Age" : 45, "Job" : "AI"}])
df = pd.concat([df, new_row])

print(df)
```

```python
import pandas as pd

data = {
    "Name" : ["Sam", "Jhon", "Bob"],
    "Age" : [20, 65, 37]
}

df = pd.DataFrame(data, index=["Person 1", "Person 2", "Person 3"])

# Add new column
df["Job"] = ["SE", "QA", "N/A"]

# Add new row
new_row = pd.DataFrame([{"Name" : "Kamal", "Age" : 45, "Job" : "AI"}], index=["Person 4"])
df = pd.concat([df, new_row])

print(df)
```

```python
import pandas as pd

data = {
    "Name" : ["Sam", "Jhon", "Bob"],
    "Age" : [20, 65, 37]
}

df = pd.DataFrame(data, index=["Person 1", "Person 2", "Person 3"])

# Add new column
df["Job"] = ["SE", "QA", "N/A"]

# Add new rows
new_rows = pd.DataFrame([{"Name" : "Kamal", "Age" : 45, "Job" : "AI"},
                         {"Name" : "Sandy", "Age" : 27, "Job" : "DB"}], 
                         index=["Person 4", "Person 5"])
df = pd.concat([df, new_rows])

print(df)
```
